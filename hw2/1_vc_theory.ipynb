# Задание 1

### a)

Максимальное количество способов поставить одну перегородку между $n$ точками на прямой равно $n+1$. То есть с одной стороны от $0$ до $n$ точек, а с другой все остальные. Таким образом:
$$m_h(n) = n + 1$$
Тогда минимальное $n$, при котором $m_h(n) < 2^n$ равно 2. Значит:
$$breakpoint = 2$$ 
Следовательно:
$$d_{VC}(h) = breakpoint - 1$$
$$d_{VC}(h) = 1$$

### b)

Найдем количество способов выбрать непустуое подмножество из $n$ точек, с помощью заданного правила. Говоря в терминах перегородок, у нас есть $n$ мест, куда мы можем поставить левую перегородку, и для каждого положения левой перегородки существует от $n$ до $1$ способа поставить правую перегородку (число способов убывает со смещением левой перегородки вправо). Кроме того, возможно выбрать пустое множество точек. Итого получаем:
$$m_h(n) = 1 + \sum_{k = 1}^n k = \frac{n(n+1)}{2} + 1$$
Отсюда несложно начти минимальное n при котором нарушается равенство $m_h(n) < 2^n$:
$$m_h(2) = \frac{2(2+1)}{2} + 1 = 4 = 2^2$$
$$m_h(3) = \frac{3(3+1)}{2} + 1 = 7 < 8 = 2^3$$
$$breakpoint = 3$$
Следовательно:
$$d_{VC}(h) = breakpoint - 1$$
$$d_{VC}(h) = 2$$
